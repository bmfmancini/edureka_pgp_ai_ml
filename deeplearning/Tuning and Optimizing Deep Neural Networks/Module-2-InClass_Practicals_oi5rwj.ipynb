```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
mnist = keras.datasets.mnist

(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train[0]
```

# Preprocessing Input Data

```python
X_train = X_train.reshape(60000, 28*28)
X_test = X_test.reshape(10000, 28*28)
```

```python
28*28
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
X_train[0]
```

```python
X_train /= 255
X_test /= 255
```

```python
X_train[0]
```

```python
Y_train
```

```python
Y_train = tf.keras.utils.to_categorical(Y_train, 10)
Y_test = tf.keras.utils.to_categorical(Y_test, 10)
```

```python
Y_train
```

```python
Y_train.shape
```

# Designing the Model

```python
model1 = Sequential()

model1.add(Dense(10, input_shape=(784, ), name='Dense1', activation='softmax'))
```

```python
model1.compile(optimizer='SGD', loss = 'categorical_crossentropy', metrics=['accuracy'])
```

```python
(60000-12000) / 64
```

```python
log_dir = 'logs/fit/model1'

tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)

training = model1.fit(X_train, Y_train, batch_size=64, epochs=50, validation_split=0.2, callbacks=tensorboard_callback)
```

```python
import random
import numpy as np
# -----------------------------
# Random prediction visualization
# -----------------------------

# Pick a random index from the test set
random_index = random.randint(0, len(X_test) - 1)

# Get the corresponding image and label
random_image = X_test[random_index]
actual_label = Y_test[random_index]

# Reshape the image for prediction (1, 784)
image_reshaped = random_image.reshape(1, 784)

# Make prediction
predicted_probs = model1.predict(image_reshaped)
predicted_label = np.argmax(predicted_probs)

# Display the image and results
plt.imshow(random_image.reshape(28, 28), cmap='gray')
plt.title(f"Actual: {actual_label} | Predicted: {predicted_label}")
plt.axis('off')
plt.show()
```

# Multilayer Perceptron Model with Dropout Layer

```python
model3 = Sequential()

model3.add(Dense(64, input_shape=(784, ), name = 'Hidden_Layer1', activation='relu'))
model3.add(Dropout(0.2))

model3.add(Dense(128, name = 'Hidden_Layer2', activation='relu'))
model3.add(Dropout(0.2))

model3.add(Dense(10, name = 'Output_layer', activation='softmax'))
```

# Introduce BatchNormalization and Regularization

```python
from tensorflow.keras import layers, regularizers

model4 = Sequential()

model4.add(Dense(64, input_shape=(784, ), name = 'Hidden_Layer1', kernel_regularizer=regularizers.l2(), activation='relu'))
model4.add(layers.BatchNormalization())
model4.add(Dropout(0.2))

model4.add(Dense(128, name = 'Hidden_Layer2', activation='relu'))
model4.add(layers.BatchNormalization())
model4.add(Dropout(0.2))

model4.add(Dense(10, name = 'Output_layer', activation='softmax'))
```

```python
model4.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
log_dir = 'logs/fit/model4'

tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)

training = model4.fit(X_train, Y_train, batch_size=128, epochs=50, validation_split=0.2, callbacks=tensorboard_callback)
```

```python
plt.plot(training.history['accuracy'])
plt.plot(training.history['val_accuracy'])
plt.legend(['train','test'], loc='upper left')
plt.show()
```

```python
plt.plot(training.history['loss'])
plt.plot(training.history['val_loss'])
plt.legend(['train','test'], loc='upper left')
plt.show()
```

# Exploring Model Checkpoint

```python
from tensorflow.keras import layers, regularizers

model7 = Sequential()

model7.add(Dense(64, input_shape=(784, ), name = 'Hidden_Layer1', kernel_regularizer=regularizers.l2(), activation='relu'))
model7.add(layers.BatchNormalization())
model7.add(Dropout(0.2))

model7.add(Dense(128, name = 'Hidden_Layer2', activation='relu'))
model7.add(layers.BatchNormalization())
model7.add(Dropout(0.2))

model7.add(Dense(10, name = 'Output_layer', activation='softmax'))
```

```python
model7.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import callbacks

lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, verbose=1)

early_stop = callbacks.EarlyStopping(patience=3, restore_best_weights=True)

checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True, verbose=1)
```

# HyperParameter Tuning

```python
from keras_tuner.tuners import RandomSearch
from keras_tuner.engine.hyperparameters import HyperParameters
```

```python
def build_model(hp):
    model = Sequential()

    model.add(Dense(64, input_shape=(784, ), name = 'Hidden_Layer1', kernel_regularizer=regularizers.l2(), activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(Dropout(0.2))
    
    model.add(Dense(128, name = 'Hidden_Layer2', activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(Dropout(0.2))
    
    model.add(Dense(10, name = 'Output_layer', activation='softmax'))

    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=hp.Choice('learning_rate', [0.01, 0.001, 0.0001])),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model
```

```python
log_dir = 'logs/fit/model'
tensorboard_cb = callbacks.TensorBoard(log_dir=log_dir)
early_stop_cb = callbacks.EarlyStopping(patience=3, restore_best_weights=True)
```

```python
tuner = RandomSearch(
        build_model,
        objective='val_accuracy',
        max_trials=5,
        executions_per_trial=1,
        directory='tuner_dir',
        project_name='Demo'
        )

```

```python
tuner.search(X_train, Y_train, epochs=5, validation_split=0.2, callbacks=[tensorboard_cb, early_stop_cb], batch_size=64)
```
