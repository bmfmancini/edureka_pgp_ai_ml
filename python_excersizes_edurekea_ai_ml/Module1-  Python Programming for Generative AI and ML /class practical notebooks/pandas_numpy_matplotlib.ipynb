##concact two data frames

```python
df1 = pd.DataFrame({'A':[1,2,3],'B':[4,5,6]})
df2 = pd.DataFrame({'A':[7,8,9],'B':[10,11,12]})
pd.concat([df1,df2], axis=0)
```

##Merge Operations

## Inner join

```python
pd.merge(df1, df2, on='A',how='inner')
```

## Left Join

```python
pd.merge(df1, df2, on='A',how='left')
```

# Right outer join

```python
pd.merge(df1, df2, on='A',how='right')
```

### Full Join

```python
pd.merge(df1, df2, on='A',how='outer')
```

# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/home/sean/car_sales.csv')
df_sample = df.head(10)
plt.xlabel('Car Make')
plt.ylabel('Revenue')
plt.title('Car Revenue by Make')
plt.plot(df_sample['Make'], df_sample['Revenue'])
```

```python
df_sample.groupby('Make').sum().plot(kind='bar')
```

# Barchart

```python
df_sample[['Units_Sold','Revenue']].set_index('Units_Sold').plot(kind='bar')
```

# piechart

```python
df_sample['Model'].value_counts().plot(kind='pie', autopct='%1.1f%%')
```

# boxplot

```python
df_sample['Revenue'].value_counts().plot(kind='box')
```
