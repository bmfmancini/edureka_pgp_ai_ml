MAP Function Excersizes

Map to upper case letters

```python
def upper_case(word) -> str:
    return word.upper()


words = ['python', 'map', 'function']
print(list(map(upper_case,words)))
```

Return letter counts

```python
words = ['apple', 'banana', 'cherry']
def count_letters(words) -> str:
    return len(words)

print(list(map(count_letters,words)))
```

Add to to input numbers

```python
numbers = [10, 20, 30, 40]
def add_10(number) -> int:
    return number + 10

print(list(map(add_10,numbers)))
```

Return the number of words

```python
def count_spaced_word(word) -> int:
    
    word_count = word.split()
    return len(word_count)

spaced_words = ["hello world", "python map function", "just one"]
print(list(map(count_spaced_word,spaced_words)))
```
