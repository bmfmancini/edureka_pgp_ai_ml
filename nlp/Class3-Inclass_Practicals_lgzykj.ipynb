# NER and Parsing

```python
# NER --> Named Entity Recognition

# Google acquired YouTube in 2006

# Rule Based approach
# Organization
# Google
# Inc
# Ltd

# Problems are still exists at NER Extraction

# Ambiguity -- > Apple (Fruit and Organization)
# Domain Specific --> New Virus Names

# Pre-Trained Models to solve NER and Parsing

# 1. SpaCy -- CNN Model + Rule Based Matching(optional)
# 2. Stanford NLP -- BiLSTM-CRF(Bi directional LSTM - Conditional Random Field(Probability))
```

```python
text = "elon musk founded SpaceX in 2002 and Tesla is based in California"
text1 = "Google acquired YouTube in 2006"
```

# NER using Spacy

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(text)
doc1 = nlp(text1)
```

```python
doc
```

```python
for ent in doc.ents:
    print('Entity: -----> ', ent.text)
    print('Label: ------> ', ent.label_)
```

```python
for ent in doc1.ents:
    print('Entity: -----> ', ent.text)
    print('Label: ------> ', ent.label_)
```

# NER using Stanza

```python
import stanza
```

```python
stanza.download('en')
```

```python
nlp_stanza =stanza.Pipeline(lang='en', processors='tokenize, ner')
```

```python
doc = nlp_stanza(text)
doc1 = nlp_stanza(text1)
```

```python
for ent in doc.ents:
    print('Entity: -----> ', ent.text)
    print('Label: ------> ', ent.type)
```

```python
for ent in doc1.ents:
    print('Entity: -----> ', ent.text)
    print('Label: ------> ', ent.type)
```
